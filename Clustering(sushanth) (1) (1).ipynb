```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime=pd.read_csv('crime_data.csv')
crime
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(crime.iloc[:,1:])
df_norm
```

```python
#creating dendrogram structure
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='average'))
```

```python
# create clusters
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc=hc.fit_predict(df_norm)
y_hc
```

```python
crime['h_clusterid']=hc.labels_
crime
```

 Kmeans clustering

```python
crime1=pd.read_csv('crime_data.csv')
crime1
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(crime.iloc[:,1:])
```

```python
# Elbow curv
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime1['Clust']=x
crime1
```

```python
crime1.iloc[:,1:5].groupby(crime1.Clust).mean()
```

```python
# Import Dataset
airline=pd.read_excel('EastWestAirlines.xlsx',sheet_name='data')
airline
```

```python
airline.info()
```

```python
airline.isnull().sum()
```

```python
air=airline.iloc[:,1:]
air
```

```python
air.values
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
tran=MinMaxScaler()
air1=pd.DataFrame(tran.fit_transform(air))
air1
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neigh=NearestNeighbors(n_neighbors=22)
nbrs=neigh.fit(air1)
distances,indices=nbrs.kneighbors(air1)
distances=np.sort(distances,axis=0)
distances=distances[:,1]
plt.plot(distances)
```

```python
dbscan=DBSCAN(eps=0.152,min_samples=12).fit(air1)
dbscan
```

```python
dbscan.labels_
```

```python
air1['cluster']=dbscan.labels_
```

```python
air1.reset_index
```

```python
df1=plt.scatter(air['Balance'],air['Qual_miles'],c=dbscan.labels_)
plt.xlabel('Balancde')
plt.ylabel('Qual')
plt.title('DBscan Cluster')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
air2=KMeans(n_clusters=3)
y_mean=air2.fit_predict(air1)
y_mean
```

```python
df2=plt.scatter(air['Balance'],air['Qual_miles'],c=y_mean)
plt.xlabel('Balance')
plt.ylabel('Qual')
plt.title('KMean Cluster')
plt.show()
```
